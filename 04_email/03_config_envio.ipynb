# Configuración y envío automáticos de emails

```python
import smtplib #pip install secure-smtplib
```

### Configuración del servidor

```python
con = smtplib.SMTP("smtp.gmail.com", 587)
```

### Iniciar conexión

```python
con.ehlo()
```

```python
con.starttls() # Conexion segura encriptada
```

```python
con.login("marc.esteve.garcia@gmail.com","nhyj eriv mgyz qwsc") #Insertar contraseña de aplicaciones
```

### Envío de emails

```python
con.sendmail("marc.esteve.garcia@gmail.com", "marc@pr0j3ct.com", "Subject: PruebaAkkodis\n\nHola, esto es una prueba")
```
