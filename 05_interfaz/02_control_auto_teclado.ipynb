# Controlar y automatizar el teclado con Python

```python
import pyautogui #pip install pyautogui
```

```python
pyautogui.size() # Devuelve el tamaño de la pantalla
```

```python
pyautogui.position()
```

```python
# Ejemplo: queremos escribir en un bloc de notas y debermos colocar el ratón en x=5469, y=550
```

```python
pyautogui.click(5469, 550) #antes especificar donde se quiere escribir
pyautogui.typewrite("Hola, esto es una prueba")
```

```python
pyautogui.click(5469, 550) 
pyautogui.typewrite(["a", "b", "left", "left", "X", "Y"])
```

```python
pyautogui.click(5469, 550)
pyautogui.typewrite(["a", "b", "left", "left", 'enter', "X", "Y"], interval=0.5)
```

```python
pyautogui.KEYBOARD_KEYS
```

```python
pyautogui.press("enter")
```

```python
pyautogui.press("f1")
```

```python
pyautogui.hotkey("ctrl", "s")
```
