# Extracción Tabla Web 

```python
import pandas as pd
import requests
from io import StringIO
```

```python
url='https://es.wikipedia.org/wiki/Poblaci%C3%B3n_mundial'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
    'Accept-Language': 'es-ES,es;q=0.9,en;q=0.8'
}

r = requests.get(url, headers=headers, timeout=20)
```

```python
r.status_code
```

```python
html_doc = r.text
```

```python
df_lista = pd.read_html(StringIO(html_doc))
```

```python
html_buffer = StringIO(html_doc)
```

```python
df_lista = pd.read_html(html_buffer)
```

# Web Scrapping

```python
from bs4 import BeautifulSoup
```

```python
html_doc = """
<html>
<head>
    <title>Web scraping example</title> 
</head> 

<body>
    <h1>Web scraping with BeautifulSoup</h1>
    <p>This is a paragraph.</p>
    <p>This is another paragraph.</p>
    <p>Third paragraph.</p>
    <a href="https://www.example.com">This is a link.</a>
</body>
</html>
"""
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
soup.title
```

```python
soup.h1
```

```python
soup.p
```

```python
soup.a
```
